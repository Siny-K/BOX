## Data Read
###  (2_d.csv : NaN값 제거한 파일)

```python
import pandas as pd
data=pd.read_csv('d_2.csv') 
data
```

### std : Centralize(중심을 0으로) & Normalization (-1 ~ 1 사이의 값 갖도록)

```python
data_CN = (data-data.mean())/(data.max()--data.min())
data_CN
```

## Variance 
### : 분산 / 변수의 흩어진 정도 

```python
data_var = data_CN.var(axis=0)  #axis가 0 이면 행을 항목으로, 1 이면 열을 항목으로
data_var
```

## Standard deviation
### : 표준편차 / 분산의 양의 제곱근 / 변수가 그 평균값을 중심으로 각 관측값이 평균에서 얼마정도 벗어나 있는지 계산 한 것

```python
data_std = data_CN.std(axis=0) 
data_std
```

## 그래프 위젯
선그래프 : 분포

점 : 비례그래프 -> 상관계수 |1|에 가깝다 (서로 대체 가능)

```python
%matplotlib inline
import pandas as pd
from ipywidgets import widgets
from IPython.display import display

class scsc():
    def __init__(self, sodf):

        self.sodf=sodf.copy()

        seldescs = widgets.SelectMultiple(options=list(self.sodf.columns),description='Selected Descriptors:',)
        display(seldescs)
        def chandler(val): 
            dfsels=pd.DataFrame(sodf[list(val)])
            return pd.scatter_matrix(dfsels, alpha=0.2, diagonal='kde')

        

        button = widgets.Button(description="Get scatter matrix")
        display(button)
        def on_button_clicked(b):
            return chandler(list(seldescs.value))
        
        button.on_click(on_button_clicked)      
```

```python
scsc(data_CN)
```

```python
data2 = data.ix[0:,3:]   # ix는 자르기 함수 , xi[열 , 행] 
data2
```

```python
CN = (data2-data2.mean())/(data2.max()--data2.min())
CN
```

## Variance - Covariance matrix
### 대각선은 자기 자신과의 비교 = Variance = 변량
### 나머지 = Covariance = 공분산 = 두 변수가 동시에 변하는 정도

```python
CNcov = CN.cov()
CNcov
```

## PCA

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA(n_components= 3)
pca.fit(np.cov(CNcov))
```

```python
pca.components_
#eigen veator 
#pc1축 : -0.4451 은 mv의 계수, 0.2415는 amw의 계수
```

```python
print (pca.explained_variance_ratio_)        # eigen value/ 숫자가 클수록 중요한 축이다.
```

```python
% matplotlib inline
import matplotlib.pyplot as plt
plt.xlabel("PCA-1")
plt.ylabel("PCA-2")
plt.title("PCA", fontsize=15)
plt.grid()
plt.plot(pca.components_[0],pca.components_[1],'r^')
```
